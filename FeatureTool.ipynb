```python
import featuretools as ft
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
train = pd.read_csv("out.csv",header = None ,names=['id', 'Address', 'Area', 'Beds', 'Full Baths', 'Half Baths',
       'Locality', 'Lot Size', 'Price'])
train = train.iloc[1:,:]

train.head(5)
```

```python
train.drop(['Half Baths'], axis=1, inplace=True)
train.drop(['id'], axis=1, inplace=True)
#train["Beds"].fillna("missing", inplace= True)
#train["Area"].fillna("missing", inplace= True)
#train["Full Baths"].fillna("missing", inplace= True)
#train["Lot Size"].fillna("missing", inplace= True)
```

```python
train.isnull().sum()
train.head(5)
```

```python
combi =  train
```

```python
# creating and entity set 'es'
es = ft.EntitySet(id = 'price')

# adding a dataframe 
es.entity_from_dataframe(entity_id = 'base', dataframe = combi, index = 'id')

es.normalize_entity(base_entity_id='base', new_entity_id='Area', index = 'Locality', 
additional_variables = ['Beds', 'Lot Size', 'Full Baths'])

print(es)
```

```python
feature_matrix, feature_names = ft.dfs(entityset=es, 
target_entity = 'base', 
max_depth = 2, 
verbose = 1, 
n_jobs = 1)
```
